```python
import pandas as pd
import os
```

```python
data_2017 = os.path.join("Resources", "2017.csv")
```

```python
data_2017 = pd.read_csv(data_2017)
```

```python
data_2017.columns = data_2017.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
data_2017.head()
```

```python
data_2017 = data_2017[["provider_state", "average_medicare_payments", "average_total_payments", "total_discharges"]]
```

```python
grouped_2017 = data_2017.groupby("provider_state").sum()
```

```python
grouped_2017["per capita total"] = grouped_2017["average_total_payments"]/grouped_2017["total_discharges"]
```

```python
grouped_2017["per capita medicare"] = grouped_2017["average_medicare_payments"]/grouped_2017["total_discharges"]
```

```python
grouped_2017.head()
```

```python
top = grouped_2017.sort_values("average_medicare_payments", ascending=False).head(n=10)
```

```python
top
```

```python
top_per_capita_total = grouped_2017.sort_values("per capita total", ascending=False).head(n=10)
```

```python
top_per_capita_total
```

```python
top_per_capita_medicare = grouped_2017.sort_values("per capita medicare", ascending=False).head(n=10)
```
