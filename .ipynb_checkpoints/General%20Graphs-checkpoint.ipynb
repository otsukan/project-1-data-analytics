```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_2017 = os.path.join("Resources", "2017.csv")
```

```python
data_2017 = pd.read_csv(data_2017)
```

```python
governors_party = os.path.join("Resources", "governors_party.csv")
```

```python
governors_party = pd.read_csv(governors_party)
```

```python
data_2017.columns = data_2017.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
data_2017["average_consumer_payments"] = data_2017["average_total_payments"] - data_2017["average_medicare_payments"]
```

```python
# Combine the data into a single dataset
data_2017 = pd.merge(data_2017, governors_party, how="inner", on=["provider_state", "provider_state"])
```

```python
data_2017.head()
```

```python
data_2017_drg = data_2017.groupby("drg_definition").sum()
len(data_2017_drg)
```

```python
data_2017 ["total_covered_charges"] = data_2017["average_covered_charges"]*data_2017["total_discharges"]
data_2017 ["total_payments"] = data_2017["average_total_payments"]*data_2017["total_discharges"]
data_2017 ["total_medicare_payments"] = data_2017["average_medicare_payments"]*data_2017["total_discharges"]
data_2017 ["total_consumer_payments"] = data_2017["average_consumer_payments"]*data_2017["total_discharges"]
```

```python

data_2017["per capita average covered charges"] = data_2017["total_covered_charges"]/data_2017["total_discharges"]
data_2017["per capita average total payments"] = data_2017["total_payments"]/data_2017["total_discharges"]
data_2017["per capita average medicare payments"] = data_2017["total_medicare_payments"]/data_2017["total_discharges"]
data_2017["per capita average consumer payments"] = data_2017["total_consumer_payments"]/data_2017["total_discharges"]
data_2017["per capita average covered - per capita average total payments"] = (data_2017["total_covered_charges"]-data_2017["total_payments"])/data_2017["total_discharges"]
```

```python
data_2017.head()
```

```python
data_2017_grouped = data_2017.groupby("state_party").sum()
```

```python
data_2017_grouped.head()
```

```python
data_2017_grouped["per capita average total payments"] = data_2017_grouped["total_payments"]/data_2017_grouped["total_discharges"]
data_2017_grouped["per capita average medicare payments"] = data_2017_grouped["total_medicare_payments"]/data_2017_grouped["total_discharges"]
data_2017_grouped["per capita average consumer payments"] = data_2017_grouped["total_consumer_payments"]/data_2017_grouped["total_discharges"]
```

```python
data_2017_grouped = data_2017_grouped["per capita average consumer payments"].reset_index()
```

```python
data_2017_grouped = data_2017_grouped.sort_values("per capita average consumer payments", ascending=False)
```

```python
data_2017_grouped
```

```python
names = []
for name in data_2017_grouped["state_party"]: 
    names.append(name)
    
```

```python
plt.figure(figsize=(15,10))
x_axis = np.arange(len(data_2017_grouped["state_party"]))

    
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, names, rotation="vertical")
plt.bar(x_axis, data_2017_grouped["per capita average consumer payments"])
plt.grid()
plt.xlabel("State")
plt.ylabel("Average Per Capita Payments ($)")
plt.title("Per Capita Average Consumer Payments by State")
plt.tight_layout()
plt.savefig("Per Capita Average Consumer Payments by State")
```

# Identify Items with Heart Failure and average cost by hospital

```python
data_2017_subset = pd.DataFrame(data_2017)
data_2017_subset
data_2017_heart_failure = data_2017_subset[data_2017.drg_definition.str.contains('HEART FAILURE')]
```

```python
data_2017_heart_failure
```

```python
pd.pivot_table(data = data_2017_heart_failure, index = ['drg_definition'], columns='provider_id', values = 'per capita average total payments',fill_value = 0)
```

```python
data_2017_heart_failure['per capita average total payments'].groupby(data_2017['drg_definition']).mean()
```

```python
data_2017_heart_failure['per capita average total payments'].groupby(data_2017['provider_id']).mean()
```

```python
data_2017_heart_failure.to_csv('heart_failure.csv')
```
