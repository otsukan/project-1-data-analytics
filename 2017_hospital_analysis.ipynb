```python
import pandas as pd
import os
```

```python
data_2017 = os.path.join("Resources", "2017.csv")
```

```python
data_2017_raw = pd.read_csv(data_2017)
```

```python
data_2017_raw.columns = data_2017_raw.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
data_2017_raw.head()
```

```python
data_2017 = data_2017_raw[["provider_state", "average_medicare_payments", "average_total_payments", "total_discharges"]]
```

```python
grouped_2017 = data_2017.groupby("provider_state").sum()
```

```python
grouped_2017["per capita total"] = grouped_2017["average_total_payments"]/grouped_2017["total_discharges"]
```

```python
grouped_2017["per capita medicare"] = grouped_2017["average_medicare_payments"]/grouped_2017["total_discharges"]
```

```python
grouped_2017.head()
```

```python
top_medicare_payments = grouped_2017.sort_values("average_medicare_payments", ascending=False).head(n=10)
```

```python
top_medicare_payments["average_total_payments"] = top_medicare_payments["average_total_payments"].map('${:,.2f}'.format)
top_medicare_payments["average_medicare_payments"] = top_medicare_payments["average_medicare_payments"].map('${:,.2f}'.format)
top_medicare_payments["per capita total"] = top_medicare_payments["per capita total"].map('${:,.2f}'.format)
top_medicare_payments["per capita medicare"] = top_medicare_payments["per capita medicare"].map('${:,.2f}'.format)
top_medicare_payments["total_discharges"] = top_medicare_payments["total_discharges"].map('{:,.2f}'.format)
top_medicare_payments
```
