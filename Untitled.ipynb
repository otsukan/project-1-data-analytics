```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_2017 = os.path.join("Resources", "2017.csv")
```

```python
data_2017 = pd.read_csv(data_2017)
```

```python
data_2017.columns = data_2017.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
data_2017
```

```python
data_2017 = data_2017[["provider_state", "total_discharges", "average_total_payments"]]
```

```python
data_2017["per capita average total payments"] = data_2017["average_total_payments"]/data_2017["total_discharges"]
```

```python
data_2017
```

```python
state_grouped = data_2017.groupby("provider_state").sum()
```

```python
state_grouped["per capita average total payments"] = state_grouped["average_total_payments"]/state_grouped["total_discharges"]
```

```python
state_grouped = state_grouped["per capita average total payments"].reset_index()
```

```python
state_grouped = state_grouped.sort_values("per capita average total payments", ascending=False)
```

```python
state_grouped
```

```python
names = []
for name in state_grouped["provider_state"]: 
    names.append(name)
```

```python
plt.figure(figsize=(15,10))
x_axis = np.arange(len(state_grouped["provider_state"]))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, names, rotation="vertical")
plt.bar(x_axis, state_grouped["per capita average total payments"])
plt.grid()
plt.xlabel("State")
plt.ylabel("Average Per Capita Payments ($)")
plt.title("Per Capita Average Total Payments by State")
plt.tight_layout()
plt.savefig("Per Capita Average Total Payments by State")
```
